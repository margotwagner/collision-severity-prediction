```python
from mpl_toolkits.basemap import Basemap
import os
import matplotlib.pyplot as plt # plotting
from mpl_toolkits.basemap import Basemap # plotting maps
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sqlite3 # read the database
import numpy as np

from sklearn.cluster import KMeans
```

```python
for dirname, _, filenames in os.walk('data/switrs.sqlite'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
with sqlite3.connect("data/switrs.sqlite/switrs.sqlite") as con:
    query = (
            "SELECT latitude, longitude "
            "FROM collisions "
            "WHERE latitude IS NOT NULL AND longitude IS NOT NULL"
        )
    
df = pd.read_sql_query(query, con)
```

```python
X = df.values
```

```python
X[0][0], X[0][1]
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(20,20))

basemap = Basemap(
    projection='gall',
    llcrnrlon = -126,   # lower-left corner longitude
    llcrnrlat = 32,     # lower-left corner latitude
    urcrnrlon = -113,   # upper-right corner longitude
    urcrnrlat = 43,     # upper-right corner latitude
)

x, y = basemap(df['longitude'].values, df['latitude'].values)

basemap.plot(x, y, 'k.', markersize=1.5)

fig.show()
```

# K Means clustering of location points

```python

kmeans = KMeans(n_clusters=4,random_state=0).fit(X)
```

```python
df['cluster'] = kmeans.labels_
```

```python
df
```

```python
def colorf(cluster_num):
    if cluster_num == 0:
        return 'g.'
    elif cluster_num == 1:
        return 'k.'
    elif cluster_num == 2:
        return 'b.'
    else:
        return 'c.'
    pass


fig = plt.figure(figsize=(20,20))

basemap = Basemap(
    projection='gall',
    llcrnrlon = -126,   # lower-left corner longitude
    llcrnrlat = 32,     # lower-left corner latitude
    urcrnrlon = -113,   # upper-right corner longitude
    urcrnrlat = 43,     # upper-right corner latitude
)

for i in range(4):
    temp_df = df[df['cluster'] == i]
    
    x, y = basemap(temp_df['longitude'].values, temp_df['latitude'].values)

    
    color = colorf(i)
    basemap.plot(x, y, color, markersize=1.5)

fig.show()

# x, y = basemap(df['longitude'].values, df['latitude'].values)
# s = 0
# z = len(df['longitude'].values)
# for lon,lat,clu in zip(df['longitude'].values, df['latitude'].values, df['cluster'].values):
#     print(s/z)
#     s+=1
#     basemap.plot(lon,lat ,marker='o',markerfacecolor = colorf(clu), markersize=1.5)

# fig.show()
```
